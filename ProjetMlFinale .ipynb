# Clustering 

**DBSCAN:** C'est un algorithme fondamentale de l'apprentissage non supervisé, il se base sur **la densité** des données pour regrouper des points similaires sans nécessiter d'étiquettes préalables.

**Les Paramètres Clés:**

 -ϵ (Epsilon): Distance maximale pour que 2 points soient voisins.

 -MinPts: Le nombre minimal de point pour former une region **dense** 

 **Points forts:**

 -Pas besoin de lui dire le nombre de groupe à l'avance. 

 -Il identifie le bruit. 

 -DBSCAN peut détecter des clusters de formes complexes.

  **Point faible:**

  -Le choix des paramètres ϵ et MinPts est sensible. 









**K-MEANS:** C'est un algorithme de l'apprentissage non supervisé aussi, il se base sur le partitionnement par centre de gravité. 

**Les Paramètres Clés:**

 -K : Nombre de classes défini dés le début. => pour déterminer K optimal on utilise soit **le score de silhouette** ou **la méthode de coude**. 

 -Les centres du K : L'algorithme nécessite une initialisation de K centres.

 **Points forts:**

 -Simplicité d'implémentation.

-Rapidité sur de grands jeux de données.

-Convergence garantie.

   **Point faible:**

   -Inefficace pour des formes complexes.

   -forte influence des outliers.
  

**ACP:** Permet de trouver une meilleure projection pour représenter fidèlement la répartition des données, l'idée centrale est de créer de nouvelles variables, appelées composantes principales, qui résument les variables initiales pour réduire la dimension du problème. 

**Principes de fonctionnement:** 
- **la variance** L'ACP se base sur la varience pour identifier les axes.

- **Nombre de composantes:** Il existe autant de composantes que de variables initiales, mais les premières expliquent généralement la majeure partie de la variabilité.

**Choix du nombre d'axes à retenir:** Soit par la variance cumulative soit par le critère de Kaiser.

**Points Forts:** 

-Implification des données.

-Visualisation facilitée et élimination du bruit.

**Points Faibles:**

-Il suppose que les relations entre les variables sont linéaires.

-Perte de données.


# Classification

**Random Forest Classifier:** est un ensemble d’arbres de décision construits via la technique de **bagging**, où chaque arbre est entraîné sur un échantillon aléatoire des données et une sélection aléatoire de caractéristiques et prédit par vote majoritaire.

**Les Paramètres Clés:** 

-n_estimators : Nombre d’arbres dans la forêt.

-max_depth : Profondeur maximale de chaque arbre.

-min_samples_split : Nombre minimum d’échantillons pour diviser un noeud.

-min_samples_leaf : Nombre minimum d’échantillons dans une feuille.

-max_features : Nombre de features à considérer à chaque split.

-bootstrap : Utiliser ou non le bootstrap.

**Points Forts:** 

-Moins sensible au surapprentissage et aux données bruitées ou incomplètes.

-Combine plusieurs arbres pour améliorer la performance par rapport à un arbrede décision simple.

-Capable de mesurer l’importance des variables.

**Points Faibles:**

-le temps d'apprentissage est plus long que celui des autres modèles.

-Random forest est plus complexe que les arbres de décision.


**Gradient Boost:** La descente de gradient est un algorithme d'optimisation itératif utilisé pour minimiser une fonction de perte. 

**Les Paramètres Clés:** 

-n_estimators : nombre d’arbres ajoutés

-learning_rate : poids de chaque nouvel arbre

-max_depth ou taille des arbres de base

**Points Forts:** 

-Très bonne performance prédictive

-Permet d’estimer l’importance des variables

**Points Faibles:**

-Plus lent à entraîner

-Sensible aux hyperparamètres

# Regression 

**Random Forest Regressor:** Est un modèle de régression qui combine plusieurs arbres de décision. Chaque arbre fait une prédiction, puis le modèle prend la moyenne de toutes les prédictions pour obtenir un résultat plus fiable.

**Les hyperparametres clés:**

-n_estimators : nombre d’arbres dans la forêt.

-max_depth : profondeur maximale de chaque arbre.

-min_samples_split : nombre minimum d’observations pour diviser un nœud.

-min_samples_leaf : nombre minimum d’observations dans une feuille.

-max_features : nombre de variables testées à chaque split.

-bootstrap : indique si les arbres sont construits sur des échantillons bootstrap.

**Points forts:**

-Bonne performance prédictive sur des données tabulaires.

-Réduit l’overfitting par rapport à un arbre unique grâce à l’agrégation par moyenne.

**Points faibles:**

-Plus lourd en calcul quand le nombre d’arbres est élevé.

-Moins interprétable qu’une régression linéaire ou qu’un arbre unique.

**XGBoost:** modèle de régression avancé basé sur le gradient boosting, utilisé pour prédire une valeur numérique en corrigeant progressivement les erreurs des arbres précédents. 
**Points forts:** 

-Il gère efficacement les données manquantes, ce qui est un avantage dans les jeux de données réels qui sont souvent incomplets.

-Il est reconnu pour sa vitesse d'exécution et sa capacité à obtenir des résultats de haute précision

**Points faibles:**

-Sensibilité aux hyperparamètres

-Nécessité d'une validation croisée

-Biais vers la majorité 



**Regression linéaire:** Consiste à trouver l'hyperplan qui minimise l'erreur entre les valeurs prédites et les valeurs réelles dans les données d’entraînement.

**Points forts:**

-Simple à comprendre et à expliquer

-Très pratique comme modèle de base pour comparer les autres approches.

**Points faibles:**

-Suppose une relation linéaire entre les variables et la cible, donc elle peut mal fonctionner si la relation réelle est non linéaire.

-Sensible aux variables corrélées entre elles



# Time series / forecasting

**Naive Forecast:**  il prédit la prochaine valeur comme la dernière observée, donc il est simple et souvent difficile à battre sur une petite série.

**Points forts:** 

-Très rapide à appliquer

-Sert de baseline pour vérifier si un modèle plus complexe apporte vraiment un gain.

**Points faibles:**

-Ignore presque toute l’information historique

-Devient vite mauvais si la série évolue régulièrement à la hausse ou à la baisse.



**ARIMA:** est un modèle statistique pour prévoir une série temporelle à partir de ses valeurs passées.

**Les hyperparamètres clès:**

-p : nombre de retards utilisés par la partie autoregressive.

-d : nombre de différenciations pour rendre la série plus stationnaire.

-q : nombre de retards des erreurs dans la partie moyenne mobile.

**Points forts:** 

-Bon pour des séries sans saisonnalité forte

-utilise l’historique de la série.

**Points faibles:**

-Le choix de p, d,q demande des essais et peut être délicat.

-Ne gère pas bien la saisonnalité si on n’utilise pas SARIMA

# ACP

```python
cluster_pca = cluster_df[[
    "ProduitPK",
    "Quantite_Produit",
    "Prix_Totale",
    "TotalDiscount",
    "PrixVenteHT",
    "PrixAchatHT"
]].copy()
```

```python
from sklearn.preprocessing import RobustScaler                # travaille avec l'IQR et la mediane c pourquoi il est robuste la standardisation simple utilise la moy et medaine donc elle peut etre influencé par les outliers 

X_num = cluster_pca.drop(columns=["ProduitPK"])

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X_num)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)                                #résumer plusieurs variables numériques en un petit nombre de nouvelles variables, tout en gardant le maximum d’information possible.
X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["ProduitPK"] = cluster_pca["ProduitPK"].values

pca_df.head()
```

```python
print("Variance expliquée par PC1 :", round(pca.explained_variance_ratio_[0] * 100, 2), "%")
print("Variance expliquée par PC2 :", round(pca.explained_variance_ratio_[1] * 100, 2), "%")
print("Variance expliquée totale :", round(pca.explained_variance_ratio_.sum() * 100, 2), "%")
```

PC1 résume à elle seule une grande partie de l’information contenue dans les variables d’origine.

PC2 ajoute encore une part importante d’information.

=> C’est un très bon résultat.

```python
plt.figure(figsize=(9, 6))
plt.scatter(pca_df["PC1"], pca_df["PC2"], alpha=0.6)
plt.axhline(0, color="grey", linewidth=1)
plt.axvline(0, color="grey", linewidth=1)
plt.title("Carte des individus - ACP")
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.2f}%)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.2f}%)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
num_cols = ["Quantite_Produit", "Prix_Totale", "TotalDiscount", "PrixVenteHT", "PrixAchatHT"]
loadings = pca.components_.T

plt.figure(figsize=(8, 8))

# cercle
circle = plt.Circle((0, 0), 1, color="blue", fill=False)
plt.gca().add_artist(circle)

# axes
plt.axhline(0, color="grey", linewidth=1)
plt.axvline(0, color="grey", linewidth=1)

# flèches des variables
for i, var in enumerate(num_cols):
    plt.arrow(
        0, 0,
        loadings[i, 0],
        loadings[i, 1],
        head_width=0.03,
        head_length=0.03,
        length_includes_head=True
    )
    plt.text(
        loadings[i, 0] * 1.12,
        loadings[i, 1] * 1.12,
        var,
        fontsize=10
    )

plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.2f}%)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.2f}%)")
plt.title("Cercle de corrélation")
plt.grid(True, alpha=0.3)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```

Le cercle de corrélation montre que la première composante principale (PC1) est surtout influencée par la quantité vendue, le prix total et la remise totale. Elle représente donc principalement l’intensité de l’activité commerciale des produits. La deuxième composante (PC2) apporte une distinction complémentaire entre les produits davantage portés par le volume et ceux plus liés à la valeur monétaire et aux remises. Enfin, les variables PrixVenteHT et PrixAchatHT sont fortement corrélées, ce qui traduit une cohérence entre coût d’achat et prix de vente.

PC1 = activité commerciale globale

PC2 = volume vs valeur/remise/prix

PrixVenteHT et PrixAchatHT = très corrélés

Prix_Totale et TotalDiscount = assez corrélés

le graphique est fiable car les 2 axes expliquent 90.47 % de la variance

# Kmeans

```python
cluster_df["Marge"] = cluster_df["PrixVenteHT"] - cluster_df["PrixAchatHT"]
```

```python
X_kmeans = cluster_df[[
    "Quantite_Produit",
    "Prix_Totale",
    "TotalDiscount",
    "PrixVenteHT",
    "PrixAchatHT",
    "Marge"
]].copy()

X_kmeans.head()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
X_kmeans_scaled = scaler.fit_transform(X_kmeans)
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
import matplotlib.pyplot as plt

inertias = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_kmeans_scaled)
    inertias.append(kmeans.inertia_)                                        # inertie : L’inertie mesure le serrage des clusters.

# Détection automatique du coude
knee = KneeLocator(K, inertias, curve="convex", direction="decreasing")     #Outil qui détecte automatiquement le point de coude sur la courbe.
k_optimal = knee.elbow

print("k optimal détecté :", k_optimal)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K, inertias, marker="o", label="Inertie")

if k_optimal is not None:
    plt.axvline(x=k_optimal, color="red", linestyle="--", linewidth=2, label=f"k optimal = {k_optimal}")

plt.title("Méthode du coude - KMeans")
plt.xlabel("Nombre de clusters k")
plt.ylabel("Inertie")  
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

L’inertie diminue lorsque le nombre de clusters augmente. 

La baisse est forte entre k = 2 et k = 5. 

Le point de coude détecté est k = 5.

Cela signifie que 5 clusters représentent un bon compromis pour segmenter les produits.

avec KneeLocator, l’algorithme peut détecter 5 comme point de rupture le plus pertinent

```python
kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_kmeans_scaled)

cluster_df["Cluster"] = clusters
cluster_df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd

cluster_names = {
    0: "Produits performants sous promotion",
    1: "Produits à faible performance commerciale",
    2: "Produits stars rentables",
    3: "Produits à très forte rotation et faible marge",
    4: "Produits premium à faible rotation"
}

pca_visu = PCA(n_components=2)
X_visu = pca_visu.fit_transform(X_kmeans_scaled)
centres_visu = pca_visu.transform(kmeans.cluster_centers_)

visu_df = pd.DataFrame(X_visu, columns=["Dim1", "Dim2"])
visu_df["Cluster"] = clusters
visu_df["Cluster_Label"] = visu_df["Cluster"].map(cluster_names)

plt.figure(figsize=(12, 7))

for c in sorted(visu_df["Cluster"].unique()):
    subset = visu_df[visu_df["Cluster"] == c]
    plt.scatter(
        subset["Dim1"],
        subset["Dim2"],
        alpha=0.6,
        label=cluster_names[c]
    )

plt.scatter(
    centres_visu[:, 0],
    centres_visu[:, 1],
    c="black",
    s=220,
    marker="X",
    label="Centres"
)

for i, (x, y) in enumerate(centres_visu):
    plt.text(x + 0.3, y + 0.3, f"C{i}", fontsize=10, color="black")

plt.title("Visualisation des clusters KMeans")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Colonnes numériques
numeric_features = ['IsPaid', 'Quantite_Produit', 'TotalDiscount']

# Colonnes catégorielles
categorical_features = ['Target_Gender', 'Target_Age', 'TypePartner']

# Pipeline numérique
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())                            #standardise les variables numériques
])

# Pipeline catégoriel
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),    #remplace les valeurs manquantes
    ('onehot', OneHotEncoder(handle_unknown='ignore'))      #transforme les variables catégorielles en colonnes numériques
])

# Préprocessing global
preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

print("Preprocessing prêt")
```

Ici on prépare les données avant d’entraîner les modèles.

Pourquoi ?

Parce que dans X

Colonnes numériques
IsPaid
Quantite_Produit
TotalDiscount
Colonnes catégorielles
Target_Gender
Target_Age
TypePartner

On ne traite pas ces deux types de colonnes de la même manière.

Ce que fait le pipeline numérique
remplace les valeurs manquantes par la médiane
standardise les valeurs
Ce que fait le pipeline catégoriel
remplace les valeurs manquantes par la modalité la plus fréquente
transforme les catégories en colonnes numériques avec OneHotEncoder
Pourquoi ColumnTransformer

Parce qu’il permet d’appliquer un traitement différent à chaque groupe de colonnes.

Parce que ton dataset est déséquilibré.

```python
#Modèle 2 — Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier

pipe_gb = Pipeline([
    ('preprocessor', preprocessor),
    ('model', GradientBoostingClassifier(
        random_state=42
    ))
])

print("Pipeline Gradient Boosting prêt")
```

Parce que c’est aussi un modèle performant.

```python
#Tuning de Random Forest avec RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

param_rf = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 5, 10, 15],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__max_features': ['sqrt', 'log2']
}

random_rf = RandomizedSearchCV(
    estimator=pipe_rf,
    param_distributions=param_rf,
    n_iter=10,
    scoring='f1',
    cv=cv_strategy,
    random_state=42,
    n_jobs=-1
)

random_rf.fit(X_train, y_train)

print("Best params RF :", random_rf.best_params_)
print("Best CV score RF :", random_rf.best_score_)
```

```python
def fix_negatives(df, name):
    numeric_cols = df.select_dtypes(include='number').columns
    report = []
    for col in numeric_cols:
        neg_count = (df[col] < 0).sum()
        if neg_count > 0:
            df[col] = df[col].abs()
            report.append(f'  {col}: {neg_count} valeurs négatives → corrigées')
    if report:
        print(f'[{name}] Corrections:')
        for line in report: print(line)
    else:
        print(f'[{name}] Aucune valeur négative ✓')
    return df

reg_df = fix_negatives(reg_df, 'reg_df')
```

#  Récupération des ventes uniquement

```python
query_ventes = text("""
    SELECT
        d.[Date]                        AS DateTransaction,
        CAST(f.Quantite_Produit AS FLOAT) AS Quantite_Produit,
        CAST(f.Prix_Totale      AS FLOAT) AS Prix_Totale
    FROM Verde_DW.dbo.FactTransaction f
    LEFT JOIN Verde_DW.dbo.DimDate d
        ON f.DateTransaction = d.DatePK
    WHERE f.TransactionType = 2
      AND d.[Date] IS NOT NULL
    ORDER BY d.[Date]
""")

ts_raw = pd.read_sql(query_ventes, engine)
ts_raw['DateTransaction']  = pd.to_datetime(ts_raw['DateTransaction'], errors='coerce')
ts_raw['Quantite_Produit'] = pd.to_numeric(ts_raw['Quantite_Produit'], errors='coerce')
ts_raw['Prix_Totale']      = pd.to_numeric(ts_raw['Prix_Totale'],      errors='coerce')
ts_raw = ts_raw.dropna()
ts_raw = ts_raw[ts_raw['DateTransaction'] >= '2024-01-01']

print(f"Période : {ts_raw['DateTransaction'].min()} → {ts_raw['DateTransaction'].max()}")
print(f"Nombre de transactions de VENTE : {len(ts_raw)}")
ts_raw.head()
```

# Agrégation hebdomadaire

```python
ts_weekly = (
    ts_raw
    .set_index('DateTransaction')
    .resample('W')
    .agg(
        Quantite_Semaine=('Quantite_Produit', 'sum'),
        CA_Semaine=('Prix_Totale', 'sum')
    )
    .reset_index()
)

# Enrichissement temporel
ts_weekly['Year']    = ts_weekly['DateTransaction'].dt.year
ts_weekly['Month']   = ts_weekly['DateTransaction'].dt.month
ts_weekly['Quarter'] = ts_weekly['DateTransaction'].dt.quarter
ts_weekly['Semaine'] = ts_weekly['DateTransaction'].dt.isocalendar().week.astype(int)

print(f"Nombre de semaines : {len(ts_weekly)}")
ts_weekly.head(10)
```

# Test de stationnarité 

```python
def test_stationnarite(serie, nom="Série"):
    result = adfuller(serie.dropna())
    print(f"=== Test ADF — {nom} ===")
    print(f"  Statistique ADF : {result[0]:.4f}")
    print(f"  p-value         : {result[1]:.4f}")
    print(f"  Seuils critiques: {result[4]}")
    if result[1] < 0.05:
        print("  ✅ Série STATIONNAIRE (p < 0.05)\n")
    else:
        print("  ⚠️  Série NON STATIONNAIRE → différenciation recommandée\n")

test_stationnarite(ts_weekly['Quantite_Semaine'], "Quantité hebdomadaire")
```

```python
n = len(ts_weekly)
split_idx = int(n * 0.8)

ts_train = ts_weekly.iloc[:split_idx].copy().reset_index(drop=True)
ts_test  = ts_weekly.iloc[split_idx:].copy().reset_index(drop=True)

print(f"Train : {len(ts_train)} semaines ({ts_train['DateTransaction'].iloc[0].date()} → {ts_train['DateTransaction'].iloc[-1].date()})")
print(f"Test  : {len(ts_test)}  semaines ({ts_test['DateTransaction'].iloc[0].date()} → {ts_test['DateTransaction'].iloc[-1].date()})")
```

# Naive Forecast

```python
last_train_value = ts_train['Quantite_Semaine'].iloc[-1]
ts_test = ts_test.copy()
ts_test['Naive_Forecast'] = last_train_value

mae_naive  = mean_absolute_error(ts_test['Quantite_Semaine'], ts_test['Naive_Forecast'])
rmse_naive = np.sqrt(mean_squared_error(ts_test['Quantite_Semaine'], ts_test['Naive_Forecast']))

print(f"=== Naive Forecast ===")
print(f"  MAE  : {mae_naive:.2f}")
print(f"  RMSE : {rmse_naive:.2f}")
```

#  Modèle ARIMA

```python
arima_model = ARIMA(ts_train['Quantite_Semaine'], order=(2, 1, 2))
arima_fit   = arima_model.fit(method_kwargs={"warn_convergence": False})

arima_forecast = arima_fit.forecast(steps=len(ts_test))
ts_test = ts_test.copy()
ts_test['ARIMA_Forecast'] = arima_forecast.values

mae_arima  = mean_absolute_error(ts_test['Quantite_Semaine'], ts_test['ARIMA_Forecast'])
rmse_arima = np.sqrt(mean_squared_error(ts_test['Quantite_Semaine'], ts_test['ARIMA_Forecast']))

print(f"=== ARIMA(2,1,2) ===")
print(f"  MAE  : {mae_arima:.2f}")
print(f"  RMSE : {rmse_arima:.2f}")
print(arima_fit.summary())
```
